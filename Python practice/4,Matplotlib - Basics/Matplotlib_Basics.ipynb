# Matplotlib

```python
# Importing libraries
import matplotlib.pyplot as plt
import numpy as np
```

# Plotting Different Functions

```python
# Plotting sine function
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
# Plotting cosine function
x = np.linspace(0,10,100)
z = np.cos(x)
plt.plot(x,z)
plt.show()
```

```python
# Plotting tangent function
x = np.linspace(0,10,100)
z = np.tan(x)
plt.plot(x,z)
plt.show()
```

```python
# Plotting parabolic function
x = np.linspace(-10,10,100)
y = x**2
plt.plot(x,y)
plt.show()
```

```python
# Plotting cubic curve
x = np.linspace(-10,10,100)
y = x**3
plt.plot(x,y)
plt.show()
```

# Plotting With Symbols

```python
# Plotting with '+' symbol
x = np.linspace(-10,10,100)
y = x**2
plt.plot(x,y,'b+') # 'b' for blue color plot.
plt.show()
```

```python
# Plotting with '--' symbol
plt.plot(x,y,'r--') # 'r' for red color plot.
plt.show()
```

```python
# plotting with '.' symbol
plt.plot(x,y,'g.') # 'g' for green color plot.
plt.show()
```

```python
# plotting with 'x' symbol
plt.plot(x,y,'rx') # 'r' for red color plot.
plt.show()
```

# Labeling The Plots

```python
# Labeling sine plot
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel("Angles")
plt.ylabel("Sine Values")
plt.title("Sine Function")
plt.show()
```

```python
# Labeling cosine plot
x = np.linspace(0,10,100)
y = np.cos(x)
plt.plot(x,y)
plt.xlabel("Angles")
plt.ylabel("Cosine Values")
plt.title("Cosine Function")
plt.show()
```

```python
# Labeling tangent plot
x = np.linspace(0,10,100)
z = np.tan(x)
plt.plot(x,z)
plt.xlabel("Angles")
plt.ylabel("Tangent Values")
plt.title("Tangent Function")
plt.show()
```

# Plotting Multiple Plots

```python
# Plotting sine & cosine & tangent functions --> 1st method
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x), 'b--')
plt.plot(x, np.cos(x), 'r+')
plt.plot(x, np.tan(x), 'g-')
plt.show()
```

```python
# Plotting sine & cosine & tangent functions --> 2nd method
x = np.linspace(0,10,100)
a = np.sin(x)
b = np.cos(x)
c = np.tan(x)
plt.plot(x,a, 'b--')
plt.plot(x,b,'r+')
plt.plot(x,c,'g-')
plt.show()
```

# Plotting Other Plots

Plotting Bar Plot

```python
# Plotting bar plot
fig = plt.figure()
axe =  fig.add_axes([0,0,1,1])
languages = ["English", "French", "Chinese", "Japanese", "Spanish", "German"]
students = [1023, 3040, 1243, 3425, 2323, 3243]
axe.bar(languages, students)
plt.show()
```

Plotting Pie Plot

```python
# Plotting pie plot
fig = plt.figure()
axe = fig.add_axes([0,0,1,1])
languages = ["English", "Spanish", "French", "Chinese", "Japanese", "German"]
students = [234, 3444, 343, 334, 3424, 324]
axe.pie(students, labels = languages, autopct = "%1.2f%%")  # '%1.2f%%' means to write
# the percentage to how much decimal places, for '1.2' the percentage
# will be say '4.22%' or '3.43%'.

plt.show()
```

Plotting Scatter Plot

```python
# Plotting scatter plot
x = np.linspace(0,10,40)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
axe2 = fig2.add_axes([0,0,1,1])
axe2.scatter(x,y, color='b')
axe2.scatter(x,z, color='r')
plt.show()
```

Plotting 3D Scatter Plot

```python
# x = 27 * np.random.random(100)
z = np.linspace(0,50,500)
y = np.sin(z)
x = np.cos(z)
fig3 = plt.figure()
axe3 = plt.axes(projection='3d')
axe3.scatter(x,y,z, c=z, cmap = "Reds")
plt.show()
```
