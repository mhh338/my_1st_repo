# Summary

```python
# 1, Importing Datasets Into Pandas DataFrame

# Importing Python libraries

# Importing California house price dataset
from sklearn.datasets import fetchopenml
housing = fetchopenml(name = "housing_prices", as_frame = True)
print("Type of housing:\n", type(housing)) # printing type of dataset
print(housing) # printing the dataset

# Converting dataset to Pandas dataframe
california_df = pd.DataFrame(housing.data, columns = housing.feature_names)
print(california_df.head()) # Printing top 5 values of dataframe
print(california_df.shape) # Printing shape of the dataframe

# Importing the data from a CSV file to a Pandas DataFrame
mnist_test_df = pd.read_csv("/content/sample_data/mnist_test.csv")
print(mnist_test_df.head())
print(mnist_test_df.shape)
print(type(mnist_test_df))

# Similar for importing other formats like XLS to a Pandas DataFrame

# 2, Exporting DataFrame to Other files (i.e., CSV, XLS, etc)

# Exporting the california_df to CSV
california_df.to_csv("/content/samples/california.csv")

# Exporting the california_df to XLS
california_df.to_excel("/content/samples/california.xls")

# 3, Creating DataFrame With Random Values

random_df = pd.DataFrame(np.random.rand(7,9))
print(random_df.head())
print(random_df.shape)
print(type(random_df))

# 4, Inspecting A DataFrame

# Number of rows and columns
print(california_df.shape)

# Printing first 7 rows
print(california_df.head(7))

# Printing last 7 rows
print(california_df.tail(7))

# 5, Information About The DataFrame

# Information
print(california_df.info())

# Number of missing values
print(california_df.isnull().sum())

# Counting values based on the labels
print(california_df.value_counts("LotArea"))

# Grouping the values based on the mean or average
print(random_df.groupby(5).mean())

# 6, Statistical Measures

# Number of values in each column
print(random_df.count())

# Mean value --> column wise
print(random_df.mean())

# Standard deviation --> column wise
print(random_df.std())

# Maximum value in each column
print(random_df.max())

# Minimum value in each column
print(random_df.min())

# All the statistical measure about the dataframe
print(random_df.describe())

# 7, Manipulating The DataFrame

# Adding a column
california_df["new-column"] = housing.target

# Removing a column
california_df.drop(columns = ["new-column"], inplace = True)

# Removing a row
california_df.drop(index = 2, inplace = True)

# Locating a row using index value
print(california_df.iloc[2]) # Using index value
print(california_df.loc[3]) # Using label

# Locating a column
print(california_df.iloc[:,3]) # Using index value only

# 8, Correlation

print(random_df.corr()) # Give correlation between different columns' value


```
