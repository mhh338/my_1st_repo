# Speed Processing of Lists Vs Numpy Arrays

```python
from time import process_time
import numpy as np
```

With Lists

```python
list = [a for a in range(20000)]
start_time = process_time()
new_list = [a**2 for a in list]
end_time = process_time()
ltime = end_time - start_time
print(ltime, type(list))
```

With NumPy

```python
np_array = np.array([a for a in range(20000)])
start_time = process_time()
new_array = np_array**2
end_time = process_time()
ntime = end_time - start_time
print(ntime, type(np_array))
```

```python
if (ltime > ntime):
  print("NumPy is faster")
else:
  print("List is faster")
```

# Dimensonal Arrays

One-dimensional Array

```python
one_d = np.array([1,2,4,4])
print(one_d)
one_d.shape
```

Multi-dimensional Array

```python
mult_d = np.array([(1,2,4,2),(2,43,2,5),(4,7,8,8)])
print(mult_d)
mult_d.shape
```

Representing in floating point data type

```python
mult_d = np.array([(1,2,4,2),(2,43,2,5),(4,7,8,8)],dtype=float)
print(mult_d)
mult_d.shape
```

Placeholder Array

```python
# Using zero function create the arrays of the given shape.
plc_hld_arr = np.zeros((5,6)) # In "(5,6)" '5' represents number of rows and '6' represents number of columns.
print(plc_hld_arr)
```

Ones Array

```python
# Creating array of only ones.
ones_arr = np.ones((6,7))
print(ones_arr)
```

Particular Value Array

```python
# Creating array of particular value.
part_val_arr = np.full((4,3), 6) # The second argument is the value of which the whole array is to be made.
print(part_val_arr)
```

Identity Matrix

```python
# Creating identity matrix
iden_mat = np.eye(6) # The argument shows there should be 6 '1' on the diagonal or in other words the matrix should be '6x6'.
print(iden_mat)
```

Random Values Array

```python
# Creating random arrays with random values.
rand_arr = np.random.random((4,5))
print(rand_arr) # The values in the arrays is in between 0 and 1.
```

Random Values Array With Specific Range

```python
# Creatinf random array with specific values range.
rand_spec_arr = np.random.randint(10,20,(4,6))
print(rand_spec_arr)
```

Evenly Spaced Array --> Required number of values

```python
# Creating evenly spaced array by specifying required number of values
even_spc_arr = np.linspace(0,20,5) # 1st and 2nd arguments are range and 3rd argument is the required number of values.
print(even_spc_arr)
```

Evenly Spaced Array --> Step size

```python
# Creating evenly spaced array by specifying step size.
stpize_arr = np.arange(0,30,6) # The 3rd argument representing the step size value.
print(stpize_arr)
```

Converting List To An Array

```python
# Converting a list to an array.
list2 = [3,4,323,42,4]
conv_arr = np.asarray(list2)
print(conv_arr)
type(conv_arr)
```
