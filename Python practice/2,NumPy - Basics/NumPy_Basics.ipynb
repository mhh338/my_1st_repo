# Speed Processing of Lists Vs Numpy Arrays

```python
from time import process_time
import numpy as np
```

With Lists

```python
list = [a for a in range(20000)]
start_time = process_time()
new_list = [a**2 for a in list]
end_time = process_time()
ltime = end_time - start_time
print(ltime, type(list))
```

With NumPy

```python
np_array = np.array([a for a in range(20000)])
start_time = process_time()
new_array = np_array**2
end_time = process_time()
ntime = end_time - start_time
print(ntime, type(np_array))
```

```python
if (ltime > ntime):
  print("NumPy is faster")
else:
  print("List is faster")
```

# Mathematical Operations On An Array

'+' sign with lists

```python
# Using '+' sign in lists
list1 = [1,2,5,4,3]
list2 = [5,37,3,7,3,6]
print(list1 + list2) # In lists '+' sign concatenate or join two lists.
```

1st method - Arithmatic operators on arrays

```python
# 1st method - Using arithmatic operators in arrays
arr1 = np.random.randint(0,10,(3,3))
arr2 = np.random.randint(10,20,(3,3))
print("Array-1:\n", arr1)
print("Array-2:\n", arr2)

print("Sum:\n", arr1 + arr2)
print("Subtraction:\n", arr1 - arr2)
print("Multiplication:\n", arr1 * arr2)
print("Division:\n", arr1 / arr2) # obtaining quotients
print("Modulus:\n", arr1 % arr2) # obtaining remainders
```

2nd method - Arithmatic functions on arrays

```python
# 2nd method - Using arithmatic functions on arrays
arr3 = np.random.randint(30,40,(3,3))
arr4 = np.random.randint(40,50,(3,3))
print("Array-3:\n", arr3)
print("Array-4:\n", arr4)

print("Sum:\n", np.add(arr3,arr4))
print("Subtraction:\n", np.subtract(arr3,arr4))
print("Multiplication:\n", np.multiply(arr3,arr4))
print("Division:\n", np.divide(arr3,arr4)) # obtaining quotients
print("Modulus:\n", np.mod(arr3,arr4)) # obtaining remainders
```

# Array Manipulation

1st method - Transpose function

```python
# 1st method - Using transpose function
array1 = np.random.randint(0,10,(2,3))
print("1st Array:\n", array1)
print("Shape of 1st array:\n", array1.shape)

trans1 = np.transpose(array1)
print("Transpose of 1st array:\n", trans1)
print("Shape of 1st transposed array:\n", trans1.shape)
```

2nd method - 'T' attribute of 'arr'

```python
# 2nd method - Using 'T' attribute of 'arr'
array2 = np.random.randint(0,10,(2,3))
print("2nd Array:\n", array2)
print("Shape of 2nd array:\n", array2.shape)

trans2 = array2.T
print("Transpose of 2nd array:\n", trans2)
print("Shape of 2nd transposed array:\n", trans2.shape)
```

Reshaping an array

```python
# Reshaping an array
norm_array = np.random.randint(0,10,(2,3))
print("Normal Array:\n", norm_array)
print("Shape of normal array:\n", norm_array.shape)

resh_array = norm_array.reshape(3,2)
print("Reshaped Array:\n", resh_array)
print("Shape of reshaped array:\n", resh_array.shape)
```
