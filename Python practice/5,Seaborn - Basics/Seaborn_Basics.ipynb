# Seaborn

```python
# Importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Visualization The Data Of Built-In Data-Sets

Total Bill vs Tip Data-Set with 'relplot'

```python
# Loading totall bill dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Setting the theme for the plot
sns.set_theme()
```

```python
# Plotting the data with 'relplot'
sns.relplot(data = tips, x = "total_bill", y = "tip",  col = "time", hue = "smoker", size = "size")
```

Iris Data-Set with 'scatterplot'

```python
# Loading iris dataset
iris = sns.load_dataset("iris")
iris.head()
```

```python
# Plotting the data with 'scatterplot' --> sepal_length vs petal_length
sns.scatterplot(x = "sepal_length", y = "petal_length", hue = "species", data = iris)
```

```python
# Plotting the data with 'scatterplot' --> sepal_length vs petal_width
sns.scatterplot(data = iris, x = "sepal_length", y = "petal_width", hue = "species")
```

Titanic Data-Set with 'countplot'

```python
# Loading titanic dataset
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
# Plotting the data with 'countplot' --> class
sns.countplot(data = titanic, x = "class")
```

```python
# Plotting the data with 'countplot' --> survived
sns.countplot(data = titanic, x = "survived")
```

Titanic Data-Set With Barplot

```python
# Plotting the data with 'barplot' --> sex vs survived
sns.barplot(data = titanic, x = "sex", y = "survived", hue = "class")
```

# External Data-Sets

```python
# Importing other data-sets
from sklearn.datasets import fetch_california_housing
house_california = fetch_california_housing()
# Converting data set to data frame
house = pd.DataFrame(house_california.data, columns = house_california.feature_names)
house["Price"] = house_california.target # Adding column
```

```python
house.head()
```

House DataFrame with Distribution Plot 'distplot' --> Price

```python
# Plotting the data-set with distribution plot
sns.distplot(house["Price"]) # To remove the warning use '.displot()' method
```

Correlation (either +ve or -ve) Plot Using Heat Map

```python
# Getting the corelation of house dataframe
correlation = house.corr()
# Constructing heat map
plt.figure(figsize=(7,7))
sns.heatmap(correlation, square = True, cbar = True, fmt = ".2f", annot = True, annot_kws = {'size':9}, cmap = "Reds")
```

```python
# 1, Importing necessary libraries

# 2, Visualization The Data Of Built-In Data-Sets

# Tips data set with 'relplot'
tips = sns.load_dataset("tips")
tips.head() # Printing 1st 5 rows of the the dataset.
sns.set_theme() # Setting the theme of the plot.
sns.relplot(data = tips, x = "total_bill", y = "tip", col = "time", hue = "smoker", size = "size") # Plotting.

# Iris Data-Set with 'scatterplot' --> sepal_length vs petal_length
iris = sns.load_dataset("iris")
iris.head()
sns.scatterplot(data = iris, x = "sepal_length", y = "petal_length", hue = "species")
# Similar for sepal_length vs petal_width

# Titanic Data-Set with 'countplot'
titanic = sns.load_dataset("titanic")
titanic.head()
sns.countplot(data = titanic, x = "class")

# Titanic Data-Set with 'barplot' --> sex vs survived
sns.barplot(data = titanic, x = "sex", y = "survived", hue = "class")

# 3, External Data-Sets

# Imorting the data-sets
from sklearn.datasets import fetch_california_housing
housing_california = fetch_california_housing()
house = pd.DataFrame(housing_california.data, columns = housing_california.feature_names) # Converting data-set to data-frame.
house["Price"] = housing_california.target # Adding column named "Price" in the data-frame with data-set.target values.
house.head()

# House DataFrame with Distribution Plot 'distplot' --> Price
sns.distplot(house["Price"])

# Correlation (either +ve or -ve) Plot Using Heat Map
correlation = house.corr()
plt.figure(figsize = (7, 7))
sns.heatmap(correlation, cbar = True, annot = True, square = True, fmt = ".1f", annot_kws = {"size": 9}, cmap = "Reds")
```
